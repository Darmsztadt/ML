 CRIM     per capita crime rate by town

 CHAS     Charles River dummy variable (= 1 if tract bounds river; 0 otherwise)
 NOX      nitric oxides concentration (parts per 10 million)
 RM       average number of rooms per dwelling
 AGE      proportion of owner-occupied units built prior to 1940
 DIS      weighted distances to five Boston employment centres
 RAD      index of accessibility to radial highways
 TAX      full-value property-tax rate per $10,000
 PTRATIO  pupil-teacher ratio by town
 B        1000(Bk - 0.63)^2 where Bk is the proportion of blacks by town
 LSTAT    % lower status of the population
 MEDV     Median value of owner-occupied homes in $1000's

```python
from sklearn.model_selection import train_test_split
```

```python
features = data[list(set(data.columns) - {"medv"})]
```

```python
features.columns
```

```python
labels=data["medv"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size=0.25)
```

```python
y_train.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
from sklearn.pipeline import Pipeline
from sklearn import set_config
set_config(display="diagram")
```

```python
land_ix,rooms_ix,age_ix=1,5,6
```

```python
data_pipeline = Pipeline([
    ('std_scaler', StandardScaler())
])

data_pipeline
```

```python
X_train.describe()
```

```python
y_train.describe()
```

```python
X_train_ready=data_pipeline.fit_transform(X_train)
```

```python
X_train_ready[0]
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lin_reg = LinearRegression()
```

```python
set_config(display="text")
lin_reg.fit(X_train_ready, y_train)
```

```python
import math
```

```python
from sklearn.metrics import mean_squared_error
```

```python
y_train_pred = lin_reg.predict(X_train_ready)
```

```python
math.sqrt(mean_squared_error(y_train, y_train_pred))
```

RMSE wynosi ok. 4.8 przy zakresie wartoci 5-50. Sabo.

```python
X_test_ready = data_pipeline.transform(X_test)
```

```python
y_preds = lin_reg.predict(X_test_ready)
```

```python
math.sqrt(mean_squared_error(y_test, y_preds))
```

```python
from sklearn.model_selection import cross_val_score
```

```python
scores = cross_val_score(lin_reg, X_train_ready, y_train, scoring = "neg_mean_squared_error", cv=10)
```

```python
rmse_scores = np.sqrt(-scores)
```

```python
rmse_scores
```

```python
rmse_scores.mean(), rmse_scores.std()
```

Wytrenowanie wicej modeli sytuacji nie poprawia.

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
f_reg = RandomForestRegressor()

f_reg.fit(X_train_ready, y_train)

np.sqrt(mean_squared_error(y_train, f_reg.predict(X_train_ready)))
```

```python
scores = cross_val_score(f_reg, X_train_ready, y_train, scoring = "neg_mean_squared_error", cv=10)
rmse_scores = np.sqrt(-scores)
rmse_scores
```

```python
rmse_scores.mean(), rmse_scores.std()
```

Tu ju lepiej.

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = [
    {'n_estimators' : [45, 50, 55], 'max_features' : [10, 11, 12, 13]}
]
```

```python
forest_reg = RandomForestRegressor()
```

```python
grid_search = GridSearchCV(forest_reg, param_grid, cv = 5, scoring = 'neg_mean_squared_error', return_train_score = True)
```

```python
grid_search.fit(X_train_ready, y_train)
```

```python
grid_search.best_params_
```

```python
np.sqrt(mean_squared_error(grid_search.best_estimator_.predict(data_pipeline.transform(X_test)), y_test))
```

```python
grid_search.best_estimator_.feature_importances_
```

```python
feature_names = list(features.columns)
feature_names
```

```python
importances = sorted(zip(grid_search.best_estimator_.feature_importances_, feature_names), reverse=True)
```

```python
importance, feature = zip(*importances)
```

```python
f_s = pd.Series(importance, index = feature)
```
